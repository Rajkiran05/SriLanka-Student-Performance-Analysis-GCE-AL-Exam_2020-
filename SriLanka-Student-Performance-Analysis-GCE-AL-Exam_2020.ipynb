```python
stream_counts = df_cleaned4['stream'].value_counts().sort_values()

colors = ['skyblue' if x != stream_counts.max() else 'orange' for x in stream_counts.values]

plt.figure(figsize=(10,6))

plt.barh(stream_counts.index, stream_counts.values, color=colors)

plt.title("Student Distribution by Stream", fontsize=14, fontweight='bold')
plt.xlabel("Number of Students")
plt.ylabel("Stream")

# Add values on bars 🔥
for i, v in enumerate(stream_counts.values):
    plt.text(v + 500, i, str(v), va='center')

plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
